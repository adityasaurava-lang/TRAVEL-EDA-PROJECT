```python
df_copy.replace("Fe Male", "Female", inplace=True)
df_copy.Gender.unique()
```

```python
#Univariate analysis for categorical columns

```

```python
for i in cateo_columns:
    plt.figure(figsize=(8,5))
    sns.countplot(x=i,data=df_copy,palette = 'muted')
    plt.title(f'Univariate analysis----{i}')
    plt.show()
```

```python
#Univariate analysis for numerical columns
```

```python
for i in numeric_columns:
    plt.figure(figsize=(8,5))
    sns.histplot(x=i,data=df_copy,kde=True,color='blue')
    plt.title(f' UNIVARIATE ANALYSIS------{i}')
    plt.show()
```
